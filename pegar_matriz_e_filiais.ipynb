```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
# acessar site sobre empresa em questão
browser = webdriver.Chrome()
time.sleep(2)
browser.get('https://www.econodata.com.br/consulta-empresa/61486650000183-DIAGNOSTICOS-DA-AMERICA-SA')
time.sleep(15)
```

```python
tabela_matrizes_filiais = browser.find_element(By.XPATH, '//div[@id="tabela-matrizes-filiais"]')
tbody = tabela_matrizes_filiais.find_element(By.TAG_NAME, 'tbody')
tr = tbody.find_elements(By.TAG_NAME, 'tr')
```

```python
tds = tr[0].find_elements(By.TAG_NAME, 'td')
len(tds)
tds_texto3 = tds[2].text.split('\n')
tds_texto3
tds_texto1 = tds[0]
tds_texto1.text
```

```python
cnpj = []
nome = []
endereco = []
cep = []
logradouro = []
estado = []
setor = []
qtd_funcionarios = []
```

```python
isNextDisabled = False
while not isNextDisabled:
    for item in range(len(tr)):
        
        # em cada tr tem 3 tds com determinado numero de textos...
        tds = tr[item].find_elements(By.TAG_NAME, 'td')
        tds_texto1 = tds[0].text.split('\n')
        tds_texto2 = tds[1].text
        tds_texto3 = tds[2].text.split('\n')

        # preencher variáveis
        cnpj.append(tds_texto1[0])
        nome.append(tds_texto1[1])
        
        # try:
        #     cep.append(tds_texto2[0])
        # except:
        #     cep.append(None)
        # try:
        #     logradouro.append(tds_texto2[1])
        # except:
        #     logradouro.append(None)
        # try:
        #     estado.append(tds_texto2[2])
        # except:
        #     estado.append(None)
        endereco.append(tds_texto2)
        setor.append(tds_texto3[0])
        qtd_funcionarios.append(tds_texto3[1])
        # time.sleep(1)
    try:
        next_button = browser.find_elements(By.XPATH, '//div[@class="next-page__click-button"]')
        next_button_class = next_button[1].get_attribute('class')
        if('last-page' in next_button_class):
            isNextDisabled=True
            break
        next_button[1].click()
        time.sleep(1)
    except:
        print(Exception)
        break
        time.sleep(1)
```

```python
rasp = {
    'CNPJ':cnpj,
    'Nome': nome,
    'Endereço': endereco,
    # 'CEP': cep,
    # 'Logradouro': logradouro,
    # 'Estado': estado,
    'Setor': setor,
    'Quantidade de funcionários': qtd_funcionarios,
}
rasp = pd.DataFrame(rasp)
rasp
```
