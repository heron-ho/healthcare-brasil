```python
# importacoes
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import openpyxl
import time
import pandas as pd
```

```python
browser = webdriver.Chrome()
browser.get('https://cnes.datasus.gov.br/pages/estabelecimentos/consulta.jsp')
time.sleep(1.5)
```

```python
cascata_estado = browser.find_element(By.XPATH, '//select[@ng-model="Estado"]')
cascata_estado.click()
```

```python
estados = cascata_estado.find_elements(By.TAG_NAME, 'option')
estado_RS = estados[21]
estado_RS.click()
```

```python
todas_cascatas = browser.find_elements(By.TAG_NAME, 'select')
len(todas_cascatas)
```

```python
# clicar em cidade com base no nome
cascata_municipio = todas_cascatas[1]
cascata_municipio.click()
time.sleep(1)
municipio = cascata_municipio.find_elements(By.TAG_NAME, 'option')

for m in range(len(municipio)):
    if('PORTO ALEGRE' in municipio[m].text):
        municipio[m].click()
```

```python
# escolher gestão com base no nome
cascata_gestao = todas_cascatas[2]
cascata_gestao.click()
gestao = cascata_gestao.find_elements(By.TAG_NAME, 'option')

for m in range(len(gestao)):
    if('MUNICIPAL' in gestao[m].text):
        gestao[m].click()
```

```python
# escolher natureza juridica com base no nome
cascata_natureza = todas_cascatas[3]
cascata_natureza.click()
natureza = cascata_natureza.find_elements(By.TAG_NAME, 'option')

for m in range(len(natureza)):
    if('ENTIDADES SEM' in natureza[m].text):
        natureza[m].click()
```

```python
#clicar em pesquisar
botao_pesquisar = browser.find_element(By.XPATH, '//button[@ng-click="pesquisaEstabelecimentos()"]')
botao_pesquisar.click()
```

```python
# definir variáveis para coleta:
nome = []
cnes = []
cnpj = []
nome_empresarial = []
natureza_juridica = []
logradouro = []
numero = []
complemento = []
bairro = []
municipio = []
uf = []
cep = []
telefone = []
dependencia = []
regional_de_saude = []
tipo_estabelecimento = []
subtipo_estabelecimento = []
gestao = []
```

```python
sus = browser.find_elements(By.XPATH, '//td[@data-title-text="Atende SUS"]')
len(sus)
```

```python
for a in range(323):
    informacoes_estabelecimento = browser.find_elements(By.XPATH, '//button[@title="Informações estabelecimento"]')
    for b in range(len(informacoes_estabelecimento)):

        #Atende sus
        sus = browser.find_elements(By.XPATH, '//td[@data-title-text="Atende SUS"]')
        # clica no botão que informa dados do estabelecimento
        informacoes_estabelecimento[b].click()
        time.sleep(2)
        # coleta os dados nas variáveis:
        inputs = browser.find_elements(By.XPATH, '//input[@class="form-control input-sm"]')
        nome.append(inputs[0].get_attribute('value'))
        cnes.append(inputs[1].get_attribute('value'))
        cnpj.append(inputs[2].get_attribute('value'))
        nome_empresarial.append(inputs[3].get_attribute('value'))
        natureza_juridica.append(inputs[4].get_attribute('value'))
        # outro.append(inputs[5].get_attribute('value'))
        logradouro.append(inputs[6].get_attribute('value'))
        numero.append(inputs[7].get_attribute('value'))
        complemento.append(inputs[8].get_attribute('value'))
        bairro.append(inputs[9].get_attribute('value'))
        municipio.append(inputs[10].get_attribute('value'))
        uf.append(inputs[11].get_attribute('value'))
        cep.append(inputs[12].get_attribute('value'))
        telefone.append(inputs[13].get_attribute('value'))
        dependencia.append(inputs[14].get_attribute('value'))
        regional_de_saude.append(inputs[15].get_attribute('value'))
        tipo_estabelecimento.append(inputs[16].get_attribute('value'))
        subtipo_estabelecimento.append(inputs[17].get_attribute('value'))
        gestao.append(inputs[18].get_attribute('value'))
        time.sleep(1)
        inputs[18].send_keys(Keys.ESCAPE)
        time.sleep(1)

    #clicar próxima pagina
    nxt_btn = browser.find_element(By.XPATH, '//a[@ng-switch-when="next"]')
    nxt_btn.click()
    time.sleep(4)
```

```python
raspagem = {
    'nome':nome,
    'CNES':cnes,
    'CNPJ':cnpj,
    'NOME EMRPESARIAL':nome_empresarial,
    'NATUREZA JURIDICA':natureza_juridica,
    'LOGRADOURO':logradouro,
    'NUMERO':numero,
    'COMPLEMENTO':complemento,
    'BAIRRO':bairro,
    'MUNICÍPIO':municipio,
    'UF':uf,
    'CEP':cep,
    'TELEFONE':telefone,
    'DEPENDENCIA':dependencia,
    'REGIONAL DE SAUDE':regional_de_saude,
    'TIPO DE ESTABELECIMENTO':tipo_estabelecimento,
    'SUBTIPO DE ESTABELECIMENTO':subtipo_estabelecimento,
    'GESTÃO':gestao
}

raspagem = pd.DataFrame(raspagem)
raspagem
```

```python
raspagem.to_excel('Entidades_Empresariais_Porto_Alegre_RS-até pag 80.xlsx')
```
