```python
# importacoes
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import openpyxl
import time
import pandas as pd
```

```python
# definindo lista de empresas:
lista_br = pd.read_excel('dados_CNES.xlsx')
cnpjs = lista_br['CNPJ']
cnpjs
```

```python
# - 1 utilização do webdriver

browser = webdriver.Chrome()
time.sleep(2)
browser.get('https://www.econodata.com.br/consulta-empresa/08807432000110-YDUQS-PARTICIPACOES-SA')
time.sleep(15)
```

```python
# remover propagandas
ad_close = browser.find_element(By.XPATH, '//img[@alt="icone fechar"]')
ad_close.click()
```

```python
# variáveis buscadas
atividade_economica = []
cnae = []
porte = []
qtd_func = []

logradouro = []
bairro = []
municipio_UF = []
cep = []
pais = []

nome_fantasia = []
data_abertura = []
natureza_juridica = []
situacao = []
tipo_unidade = []
opcao_pelo_simples = []
enquadramento_de_porte = []
capital_social = []
opcao_pelo_mei = []
situacao_especial = []
```

```python
# loop para buscar empresas por cnpjs
for cnpj in cnpjs:
    
    try:
        #encontra input e manda cnpj
        input_cnpj = browser.find_element(By.XPATH, '//input[@id="input-generico"]')
        input_cnpj.click()
        #envia cnpj e dar enter
        input_cnpj.send_keys(cnpj)
        input_cnpj.send_keys(Keys.ENTER)
        time.sleep(10)
    
        #procurar primeira empresa que aparece e clicar nela
        tabela_resultado = browser.find_element(By.XPATH, '//div[@id="tabela-resultado-busca-empresa"]')
        tbody = tabela_resultado.find_element(By.TAG_NAME, 'tbody')
        tr = tbody.find_elements(By.TAG_NAME, 'tr')
        td = tr[0].find_elements(By.TAG_NAME, 'td')
        a = td[0].find_element(By.TAG_NAME, 'a')
        a.click()
        time.sleep(10)
        
    except:
        time.sleep(1)
        atividade_economica.append(None)
        cnae.append(None)
        porte.append(None)
        qtd_func.append(None)
        logradouro.append(None)
        # bairro.append('não encontrado')
        municipio_UF.append(None)
        # cep.append('não encontrado')
        # pais.append('não encontrado')
        nome_fantasia.append(None)
        data_abertura.append(None)
        natureza_juridica.append(None)
        situacao.append(None)
        situacao_especial.append(None)
        tipo_unidade.append(None)
        opcao_pelo_simples.append(None)
        enquadramento_de_porte.append(None)
        capital_social.append(None)
        opcao_pelo_mei.append(None)
        browser.back()
        time.sleep(5)
        continue # vai para a próxima empresa
        
        
    # se conseguir procurar cnpj e clicar na primeira empresa, guarda o que queremos nas variáveis
    div_detalhes = browser.find_element(By.XPATH, '//div[@class="grid md:grid-cols-2 gap-6 my-2"]')
    div_detalhes_texto = div_detalhes.text.split('\n') # todos textos
    # div_detalhes_texto
    
    #atividade econômica
    atividade_economica.append(div_detalhes_texto[1])
    #cnae
    cnae.append(div_detalhes_texto[4])
    #porte
    porte.append(div_detalhes_texto[6])
    #quantidade de funcionários
    qtd_func.append(div_detalhes_texto[8])
    
    
    #localização
    div_localizacao = browser.find_element(By.XPATH, '//div[@itemprop="address"]')
    divs_textos = div_localizacao.find_elements(By.XPATH, '//div[@class="inline"]')
    try:
        logradouro.append(browser.find_element(By.XPATH, '//div[@itemprop="streetAddress"]').text)
        municipio_UF.append(browser.find_element(By.XPATH, '//div[@itemprop="addressLocality"]').text)
    except:
        logradouro.append(None) 
        municipio_UF.append(None)
    
    #dados receita federal
    dados_rf = browser.find_element(By.XPATH, '//div[@id="receita-section"]').find_element(By.XPATH, '//div[@class="grid md:grid-cols-2 gap-4"]')
    dados_rf_texto = dados_rf.text.split('\n')
    dados_rf_texto
    
    nome_fantasia.append(dados_rf_texto[5]) 
    data_abertura.append(dados_rf_texto[7])
    natureza_juridica.append(dados_rf_texto[9])
    situacao.append(dados_rf_texto[11])
    situacao_especial.append(dados_rf_texto[13])
    tipo_unidade.append(dados_rf_texto[15])
    opcao_pelo_simples.append(dados_rf_texto[17])
    enquadramento_de_porte.append(dados_rf_texto[19])
    capital_social.append(dados_rf_texto[21])
    opcao_pelo_mei.append(dados_rf_texto[23])
    
    
```

```python
# adicionar colunas novas à lista de empresas
lista_br['Atividade Econômica'] = atividade_economica + [None] * (len(lista_br) - len(atividade_economica))
lista_br['Código CNAE'] = cnae + [None] * (len(lista_br) - len(cnae))
lista_br['Porte'] = porte  + [None] * (len(lista_br) - len(porte))
lista_br['Quantidade de Funcionários'] = qtd_func + [None] * (len(lista_br) - len(qtd_func))
lista_br['Logradouro'] = logradouro + [None] * (len(lista_br) - len(logradouro))
# lista_br['Bairro'] = bairro + [None] * (len(lista_br) - len(bairro))
lista_br['Município/UF'] = municipio_UF + [None] * (len(lista_br) - len(municipio_UF))
# lista_br['CEP'] = cep + [None] * (len(lista_br) - len(cep))
# lista_br['País'] = pais + [None] * (len(lista_br) - len(pais))
lista_br['Nome Fantasia'] = nome_fantasia + [None] * (len(lista_br) - len(nome_fantasia))
lista_br['Data Abertura'] = data_abertura + [None] * (len(lista_br) - len(data_abertura))
lista_br['Natureza Jurídica'] = natureza_juridica + [None] * (len(lista_br) - len(natureza_juridica))
lista_br['Situação'] = situacao + [None] * (len(lista_br) - len(situacao))
lista_br['Situação Especial'] = situacao_especial + [None] * (len(lista_br) - len(situacao_especial))
lista_br['Tipo Unidade'] = tipo_unidade + [None] * (len(lista_br) - len(tipo_unidade))
lista_br['Opção pelo Simples'] = opcao_pelo_simples + [None] * (len(lista_br) - len(opcao_pelo_simples))
lista_br['Enquadramento de Porte'] = enquadramento_de_porte + [None] * (len(lista_br) - len(enquadramento_de_porte))
lista_br['Capital Social'] = capital_social + [None] * (len(lista_br) - len(capital_social))
lista_br['Opção pelo MEI'] = opcao_pelo_mei + [None] * (len(lista_br) - len(opcao_pelo_mei))
lista_br
```
